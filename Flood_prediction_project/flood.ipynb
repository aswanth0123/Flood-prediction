## importing necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
```

## data loading


```python
data=pd.read_csv("kerala.csv")
data.drop("SUBDIVISION",axis=1,inplace=True)
x_input=data.drop("FLOODS",axis=1)
x_input = x_input.drop("YEAR",axis = 1)
y_output=data["FLOODS"]
year = data["YEAR"]

print(x_input)
```

```python
ldr=LabelEncoder()
y_output=ldr.fit_transform(y_output)

x_input=MinMaxScaler().fit_transform(x_input)
x_input = pd.DataFrame(x_input)
x_input = pd.concat([year,x_input],axis=1)
x_input.rename(columns={"YEAR":13},inplace=True)

print(x_input)


# x_input.to_csv("created_data.csv",index=False)   
```

### the input data is now transfered into created_data.csv

```python
fetched_data = np.loadtxt("created_data.csv",delimiter=",")   # this is the input_data and y_output is the output data


# print(y_output[:10])

# print(fetched_data)
# print(y_output)
```

### Splitting train and test data

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(fetched_data,y_output,test_size=0.08)

print(x_train)
```

# Model selection

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()

model.fit(x_train,y_train)
```

### Testing

```python
prediction = model.predict(x_test)
print(prediction)

print(y_test)
```

## Accuracy score 

```python
from sklearn.metrics import accuracy_score

score = accuracy_score(prediction,y_test)

print("accuracy score is", score)
```
